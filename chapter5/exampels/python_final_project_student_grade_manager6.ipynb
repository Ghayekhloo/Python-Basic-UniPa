## Part 6 - Step 3: Build a Score Classification Function

We create a function that returns a label based on the score.


````python
def classify_score(score):
    if score >= 90:
        return "Excellent"
    elif score >= 75:
        return "Good"
    elif score >= 60:
        return "Pass"
    else:
        return "Needs improvement"

print(classify_score(95))
print(classify_score(81))
print(classify_score(63))
print(classify_score(40))
````

## Part 7 - Step 4: Build Major-Based Advice

Now we create stronger advice based on the student's major.

This is important because your students come from different academic fields.


````python
def major_advice(major):
    if major == "Supply chain management":
        return "Focus on logistics, inventory systems, planning, and analytical decision making."
    elif major == "Communication Design":
        return "Build a strong visual portfolio and improve storytelling, branding, and presentation skills."
    elif major == "Marine ecology":
        return "Strengthen field observation, marine data analysis, and environmental reporting."
    elif major == "Electronics Engineering":
        return "Practice circuits, sensors, embedded systems, and technical programming."
    elif major == "Biodiversity":
        return "Develop skills in conservation analysis, species observation, and scientific communication."
    elif major == "Vegetation science":
        return "Improve plant classification, ecosystem mapping, and environmental interpretation."
    elif major == "Mechanical Engineering" or major == "Mechanical engineer":
        return "Work on mechanics, engineering design, CAD, and problem solving."
    elif major == "ecological Modeling":
        return "Use Python, mathematics, statistics, and simulation to study ecosystems."
    else:
        return "Combine innovation, manufacturing systems, and aerospace thinking with strong teamwork."

print(major_advice("Marine ecology"))
print(major_advice("Communication Design"))
print(major_advice("Mechanical Engineering"))
print(major_advice("Mechanical manufacturing management and aerospace innovation"))
````

## Part 8 - Step 5: Build a Scholarship Decision Function

Here we use multiple conditions.

This function uses:
- score
- major
- performance level


````python
def scholarship_status(major, score):
    if score >= 90:
        return "Scholarship candidate"
    elif major in ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"] and score >= 85:
        return "Strong candidate for research support"
    elif major in ["Mechanical Engineering", "Mechanical engineer", "Electronics Engineering"] and score >= 80:
        return "Strong candidate for technical project support"
    elif score >= 70:
        return "Good standing"
    else:
        return "Needs more improvement before scholarship consideration"

print(scholarship_status("Marine ecology", 88))
print(scholarship_status("Mechanical Engineering", 82))
print(scholarship_status("Communication Design", 74))
````

## Part 9 - Step 6: Read Data and Build Report Lines

Now we combine:
- files
- loops
- strings
- functions
- conditions


````python
report_lines = []

with open("student_grades.txt", "r") as file:
    for line in file:
        name, major, score_text = line.strip().split(",")
        score = int(score_text)

        level = classify_score(score)
        advice = major_advice(major)
        scholarship = scholarship_status(major, score)

        report = f"{name} | {major} | Score: {score} | Level: {level} | {scholarship} | Advice: {advice}"
        report_lines.append(report)

for item in report_lines:
    print(item)
````

## Part 10 - Step 7: Save the Final Report

Now we save the report into a new text file.


````python
with open("final_student_report.txt", "w") as file:
    file.write("Final Student Grade Report\n")
    file.write("==========================\n")

    for item in report_lines:
        file.write(item + "\n")

print("Report saved successfully.")

with open("final_student_report.txt", "r") as file:
    print(file.read())
````

## Part 11 - Using Lists in the Project

Sometimes we want to work with lists directly after reading a file.


````python
student_names = []
student_majors = []
student_scores = []

with open("student_grades.txt", "r") as file:
    for line in file:
        name, major, score_text = line.strip().split(",")
        student_names.append(name)
        student_majors.append(major)
        student_scores.append(int(score_text))

print(student_names)
print(student_majors)
print(student_scores)
````

### Stronger example: print custom messages for each student


````python
for i in range(len(student_names)):
    name = student_names[i]
    major = student_majors[i]
    score = student_scores[i]

    if score >= 85:
        print(f"{name} from {major} is one of the strongest students in the class.")
    elif score >= 70:
        print(f"{name} from {major} is doing well and can improve even more.")
    else:
        print(f"{name} from {major} should review the basics and practice more.")
````

## Part 12 - `while` Loop Review Inside the Project

We can also process the same data with a `while` loop.


````python
i = 0

while i < len(student_names):
    name = student_names[i]
    major = student_majors[i]
    score = student_scores[i]

    print(f"{i+1}. {name} | {major} | {score}")
    i += 1
````

## Part 13 - Strong Major-Based Summary Statistics

Now let us count how many students belong to broad academic categories.


````python
science_count = 0
engineering_count = 0
other_count = 0

science_majors = ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"]
engineering_majors = [
    "Electronics Engineering",
    "Mechanical Engineering",
    "Mechanical engineer",
    "Mechanical manufacturing management and aerospace innovation"
]

for major in student_majors:
    if major in science_majors:
        science_count += 1
    elif major in engineering_majors:
        engineering_count += 1
    else:
        other_count += 1

print("Science / Environment students:", science_count)
print("Engineering / Technology students:", engineering_count)
print("Other students:", other_count)
````

## Part 14 - Build Small Helper Functions

Good projects use small helper functions.
This keeps code more organized and easier to read.


````python
def is_science_major(major):
    science_majors = ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"]
    return major in science_majors

def is_engineering_major(major):
    engineering_majors = [
        "Electronics Engineering",
        "Mechanical Engineering",
        "Mechanical engineer",
        "Mechanical manufacturing management and aerospace innovation"
    ]
    return major in engineering_majors

print(is_science_major("Marine ecology"))
print(is_science_major("Communication Design"))
print(is_engineering_major("Mechanical Engineering"))
print(is_engineering_major("Biodiversity"))
````

## Part 15 - Final Project Function

Now we put the project into one main function.
This is the cleanest version of the final project.


````python
def classify_score(score):
    if score >= 90:
        return "Excellent"
    elif score >= 75:
        return "Good"
    elif score >= 60:
        return "Pass"
    else:
        return "Needs improvement"

def major_advice(major):
    if major == "Supply chain management":
        return "Improve logistics, planning, and data-based decision making."
    elif major == "Communication Design":
        return "Build stronger branding, portfolio, and visual communication skills."
    elif major == "Marine ecology":
        return "Improve marine fieldwork, environmental observation, and scientific reporting."
    elif major == "Electronics Engineering":
        return "Practice circuits, programming, and smart device development."
    elif major == "Biodiversity":
        return "Develop conservation analysis and ecological communication skills."
    elif major == "Vegetation science":
        return "Strengthen plant science and ecosystem mapping skills."
    elif major == "Mechanical Engineering" or major == "Mechanical engineer":
        return "Build mechanics, CAD, simulation, and engineering design confidence."
    elif major == "ecological Modeling":
        return "Use coding, statistics, and mathematical models in ecology."
    else:
        return "Develop manufacturing innovation and aerospace systems thinking."

def scholarship_status(major, score):
    if score >= 90:
        return "Scholarship candidate"
    elif major in ["Marine ecology", "Biodiversity", "Vegetation science", "ecological Modeling"] and score >= 85:
        return "Research support candidate"
    elif major in ["Mechanical Engineering", "Mechanical engineer", "Electronics Engineering"] and score >= 80:
        return "Technical project candidate"
    elif score >= 70:
        return "Good standing"
    else:
        return "Needs improvement"

def build_student_grade_report(input_file, output_file):
    with open(input_file, "r") as file:
        lines = file.readlines()

    with open(output_file, "w") as file:
        file.write("Student Grade Manager Report\n")
        file.write("============================\n")

        for line in lines:
            name, major, score_text = line.strip().split(",")
            score = int(score_text)

            level = classify_score(score)
            advice = major_advice(major)
            scholarship = scholarship_status(major, score)

            file.write(f"{name} | {major} | {score} | {level} | {scholarship} | {advice}\n")

build_student_grade_report("student_grades.txt", "student_grade_manager_report.txt")

with open("student_grade_manager_report.txt", "r") as file:
    print(file.read())
````
