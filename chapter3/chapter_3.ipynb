## 1. List basics

```python
# A list stores multiple values in a single variable
# Syntax: variable = [item1, item2, ...]

names = ['Sara', 'Marco', 'Claudia']
print(names)
```

```python
numbers = [1, 2, 3, 4]
print(numbers)
```

You can access items by **index** (position). Indexing starts from 0.

```python
letters = ['z', 'a', 'b', 'c']
print(letters[1])   # second item: 'a'

numbers = [1, 2, 3, 4]
print(numbers[0])   # first item: 1

print(names[1])    # 'Marco'
```

### Changing, adding, and removing items

```python
# Change an item by index
names = ['Sara', 'Marco', 'Claudia']
names[2] = 'Maryna'
print(names)
```

```python
# Add item at the end with append()
names = ['Sara', 'Marco', 'Claudia']
names.append('Maryna')
print(names)
```

```python
# Remove an item by value with remove()
names = ['Sara', 'Marco', 'Claudia']
names.remove('Sara')
print(names)
```

### Example: build a shopping list with input

```python
shopping = ['Banana']
item1 = input('Enter a list item: ')
item2 = input('Enter a list item: ')
item3 = input('Enter a list item: ')

shopping.append(item1)
shopping.append(item2)
shopping.append(item3)

print(shopping)
```

## 2. `for` loops and `range()`

```python
# Example 1: repeat a message 5 times
for i in range(5):
    print('hello')
```

```python
# Example 2: show the counter value
for i in range(3):
    print(i)
```

### for loop over a list

```python
# Simple for-loop over a list of names
names = ['Sara', 'Marco', 'Claudia']

for name in names:
    print(name)
```

```python
# Another example: fruit list
fruit = ['apple', 'banana']

for f in fruit:
    print(f)
```

```python
# Example: reading sensor items 5 times
for i in range(5):
    print('Reading sensor item...')
```

### New example: sum and average of numbers in a list

```python
scores = [18, 25, 30, 27]

# compute total
total = 0
for score in scores:
    total += score

average = total / len(scores)
print('Total score:', total)
print('Average score:', average)
```

## 3. for loop with limited attempts (password example)

```python
# We allow the user 3 attempts to enter the correct password.
password = '1234'

for attempt in range(3):
    user = input('Enter password: ')

    if user == password:
        print('Access granted')
        break
    else:
        print('Try again.')
```

## 4. `while` loops

```python
# Example 1: simple counter
count = 2
while count < 6:
    print(count)
    count += 1
```

```python
# Example 2: password with while (maximum 3 attempts)
password = '1234'
attempt = 0

while attempt < 3:
    user = input('Enter password: ')

    if user == password:
        print('Access granted')
        break
    else:
        print('Try again.')
        attempt += 1
```

### Example 3: guess the secret number

```python
secret = 7
guess = 0

while guess != secret:
    guess = int(input('Guess the number: '))

    if guess == secret:
        print('You win!')
    else:
        print('Not correct, try again.')
```

### New example: while loop to build a dynamic list

```python
# The user can keep adding items until they type 'done'
items = []

while True:
    item = input("Enter an item (or 'done' to finish): ")

    if item == 'done':
        break

    items.append(item)

print('Your final list:', items)
```

### Summary
- A **list** stores multiple values in one variable.
- `for` loops help repeat an action for each item or a fixed number of times.
- `while` loops repeat as long as a condition is `True`.
- Use loops with lists and input to build interactive programs.
